```python
import numpy as np
import matplotlib.pyplot as plt
from  matplotlib import ticker
from matplotlib import rcParams
import seaborn as sns
```

Solucion de una matriz tridiagonal (T) : Tx = b
Dado que nuestra estrategia fue tener a las diagonales como variables:
a (subdiagonal), b (diagonal principal), c (superdiagonal).
Entonces para el d2 vemos como funciona paso por paso:

```python
def resolverU(b, c, y):
    n = len(y)
    solucion = list()
    i = n - 1
    while i >= 0:
        if b[i] == 0:
            print("Matriz no tiene solucin")
        if i == n-1:
            solucion.insert(0, y[i] / b[i])
        else:
            solucion.insert(0, (y[i] - c[i]*solucion[0]) / b[i])
        i -= 1
    return solucion

def resolverL(a, d):
    n = len(d)
    y = list()
    i = 0
    while i < n:
        if i == 0:
            y.append(d[0])
        else:
            y.append((d[i] - a[i]*y[-1]))
        i += 1
    return y


def encontrarL(a, b, c):
    n = len(c)
    L = [0]*n
    for i in range(1, n):
        if a[i] == 0:
            continue
        m = a[i] / b[i-1]
        L[i] = m
        b[i] = b[i] - c[i-1] * m
    return L
```

Retornamos L que es una lista con los multiplicadores (m).
Entonces, a va a hacer la matriz de los multiplicadores.

```python
n = 101

a = [1] * n
a[0] = 0

b = [-2] * n

c = [1] * n
c[-1] = 0


a = encontrarL(a, b, c)

print("multiplicadores de L:", a)
```

```python
d2 = [4/(n**2)] * n

y = resolverL(a, d2)
print("la solucion de y es:", y)
```

Dan negativos por que estamos

```python
solucion = resolverU(b, c, y)
print("la solucion de d es:", solucion)
```

Ahora grafico y veo si tiene sentido

```python

ax = plt.plot( solucion, label="ecuacion (b)");
# Personalizar los ejes x e y
plt.xticks(np.arange(0, 102, 10))  # Marcar cada 10 en el eje x
plt.yticks(np.arange(-1, 0.25, 0.1))  # Marcar cada 0.1 en el eje y

plt.title("Funcin de U respecto a los diferentes i")
plt.xlabel("Tamao de muestras")
plt.ylabel("Valores que toma U")
plt.legend()
plt.grid(True)
plt.show()

```
